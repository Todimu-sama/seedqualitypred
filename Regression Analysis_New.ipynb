## MACHINE LEARNING - MULTIPLE LINEAR REGRESSION

```python
# importing scaling, data splitting and accuracy libraries
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error
```

```python
# Getting our predictor and target variables
# X predictor variable
# y target variable

X = pd.DataFrame(data, columns= ['Months','Temp','RH', 'DI', 'Protein', 'Oil', 'Starch', 'Ash', 'DM', 'Crude Fib'])
y = pd.DataFrame(data, columns= ['MC', 'BC', 'FC', 'SV'])
```

```python
# Viewing the predictor variables
X
```

```python
# Viewing the target variables
y
```

```python
# scaling our data using Standard Scaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
# importing the regression modules from scikit-learn
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
```

```python
#Linear Regression analysis
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
# checking accuracy of the linear regression models
linRegPred = lin_reg.predict(X_test)
print('Linear Regression prediction accuracy : {}'.format(lin_reg.score(X_test, y_test)))
```

```python
linRegPred
```

```python
Pred = pd.DataFrame(linRegPred, columns= ['MC_pred', 'BC_pred', 'FC_pred', 'SV_pred'])
Pred
```

```python
# Lasso regression analysis
lasso = Lasso(alpha=0.01)
lasso.fit(X_train, y_train)
```

```python
# checking the accuracy for the Lasso regression models 
LassoRegPred = lasso.predict(X_test)
print('Lasso Regression prediction accuracy : {}'.format(lasso.score(X_test, y_test)))
```

```python
Pred = pd.DataFrame(LassoRegPred, columns= ['MC_pred', 'BC_pred', 'FC_pred', 'SV_pred'])
Pred
```

```python
# predicting for all values in the dataset
Pred_new = lin_reg.predict(X_scaled)
```

```python
Pred = pd.DataFrame(Pred_new, columns= ['MC_pred', 'BC_pred', 'FC_pred', 'SV_pred'])
Pred
```

```python
data['MC_pred'] = Pred['MC_pred']
data['BC_pred'] = Pred['BC_pred']
data['FC_pred'] = Pred['FC_pred'].astype('int')
data['SV_pred'] = Pred['SV_pred'].astype('int')
```

## Creating Pipelines for building the API

```python
# importing pipeline module
from sklearn.pipeline import Pipeline
```

```python
pipe_lin = Pipeline([('Scaler', StandardScaler()), ('Regressor', LinearRegression())])
```

```python
# splitting our dataset into training and testing
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
pipe_lin.fit(X_tr, y_tr)
```

```python
pipe_lin.predict(X_te)
```

```python
# saving the model as a pickle file
with open('regression_model.pkl', 'wb') as f:
    pickle.dump(pipe_lin, f)
```
